```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
data = pd.read_csv("../Data/tenyear.csv", index_col=[0])
df = pd.DataFrame(data)
df.head()
```

```python
cleaned_df = df.drop(['artists','explicit', 'mode','release_date','name','key', 'popularity'], axis=1, inplace=False)
cleaned_df = cleaned_df.set_index(["id"])
cleaned_df.head()
```

```python
five_years = cleaned_df.loc[cleaned_df['year'] > 2015]
five_years
```

```python
five_years = five_years.drop(["year"], axis=1)
five_years
```

```python
five_years.to_csv("../Data/fiveyears.csv")
```
